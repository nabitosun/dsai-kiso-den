# **Advanced data types**

### **Listas**

 List (リスト型)

**Description:** A collection used to handle many pieces of data together in numbered "boxes" where the order matters.  

**Syntax:** `variable_name = [item1, item2, ...].`  

**Feature:** It is a mutable (changeable) collection where you can specify a number (index) starting from 0 to retrieve or change data.

```python
# Example (Easy)
fruits = ["apple", "lemon", "grape"]
print(fruits[0])  # Output: apple

# Example (Hard)
data = [100, "Kyoto", True, [1, 2]]
data[1] = "Osaka" # Lists are mutable (changeable)
data.append(3.14) # You can add items
print(data)
```

```python
## OPERACIONES
# A. Crear y acceder
frutas = ["manzana", "pera"]
primera = frutas[0]  # Acceder por índice

# B. Añadir elementos
frutas.append("naranja")      # Añadir al final
frutas.insert(1, "uva")       # Insertar en una posición específica

# C. Eliminar elementos
frutas.remove("pera")         # Eliminar por nombre del valor
borrado = frutas.pop(0)       # Eliminar por índice (y devuelve el valor)

# D. Otras operaciones
cantidad = len(frutas)        # Tamaño de la lista
existe = "uva" in frutas      # Verificar si existe un elemento (True/False)
frutas.sort()                 # Ordenar la lista permanentemente

print("Lista final:", frutas)
```

### **Tuplas**
Tuple (タプル型)

**Description:** A collection specific to Python that acts like a list but is designed to be permanent.  
**Syntax:** `variable_name = (item1, item2, ...).`  
**Feature:** It is immutable (cannot be changed), uses less memory than a list, and generally allows for faster processing.

```python
# Example (Easy)
t = (12345, 54321, "hello!")
print(t[0])  # Output: 12345

# Example (Hard)
# Often used for fixed data like coordinates or returning multiple values from a function
coordinates = (35.0, 135.0)
# coordinates[0] = 40.0  # This would cause an ERROR because tuples are immutable
lat, lon = coordinates   # Unpacking a tuple
print(f"Lat: {lat}, Lon: {lon}")

print(coordinates)
print(lat)
print(lon)
```

```python
## OPERACIONES
# A. Crear y desempaquetar
punto = (10, 20, 30)
x, y, z = punto             # Tuple Unpacking

# B. Búsqueda y conteo
veces = punto.count(10)     # Cuenta cuántas veces aparece el 10
posicion = punto.index(20)  # Encuentra la posición del valor 20

# C. Combinación (Crea una nueva tupla, no modifica la original)
nueva_tupla = punto + (40, 50)

print("Tupla original:", punto)
print("Nueva tupla combinada:", nueva_tupla)
```

### **Sets**
Set (集合型)

**Description:** A collection used for storing unique items where the sequence does not matter.  

**Syntax:** `variable_name = {item1, item2, ...}` or `set(list_data).  `

**Feature:** It automatically deletes duplicate content and does not maintain order, so you cannot retrieve items by a specific position.

```python
sofianavidena = ["s","o","f","i","a","fa","la","la","la","la","la","la","la","la"]
print(sofianavidena)
set(sofianavidena)
print(set(sofianavidena))
```

```python
basket = {'apple', 'orange', 'apple', 'pear'}
print(basket)  # Output: {'orange', 'pear', 'apple'} (order varies)

# Using sets to perform mathematical set operations
a = set('abracadabra')
b = set('alacazam')
print(a - b)  # Letters in a but not in b (Difference)
print(a | b)  # Letters in a or b or both (Union)

c = set('colombia')
d = set('colombina')
print(d - c)
print(c | d)
print(c & d)
```

```python
## OPERACIONES
# A. Crear y añadir
colores = {"rojo", "azul"}
colores.add("verde")         # Añadir un elemento
colores.add("rojo")          # No hace nada (ya existe)

# B. Eliminar
colores.discard("azul")      # Elimina sin dar error si no existe
# colores.remove("amarillo") # Esto DARÍA ERROR si no existe

# C. Operaciones matemáticas entre sets
set_a = {1, 2, 3}
set_b = {3, 4, 5}

union = set_a | set_b        # Elementos en ambos {1, 2, 3, 4, 5}
interseccion = set_a & set_b  # Solo los comunes {3}
diferencia = set_a - set_b    # Solo los que están en A pero no en B {1, 2}

print("Set de colores:", colores)
print("Intersección:", interseccion)
```

 ### **Diccionarios**
Dictionary (辞書型)

**Description:** Known as an "associative array" in other languages, it stores data in pairs.  

**Syntax:** `variable_name = {key1: value1, key2: value2, ...}`.  

**Feature:** You specify a unique "key" to retrieve its corresponding "value," instead of using a number index and the elements can be entered in any order. Nota: Aunque los valores de un diccionario pueden ser casi cualquier cosa (incluyendo tuplas), las llaves (keys) deben ser de un tipo que no cambie.

```python
# Example (Easy)
prices = {"apple": 100, "mango": 150}
print(prices["apple"])

# Example (Hard)
# Nested dictionaries are common for structured data
students = {
    "22001": {"math": 90, "english": 85},
    "22002": {"math": 100, "english": 68}
}
# Accessing nested data
print(f"Student 22002 Math Grade: {students['22002']['math']}")
```

```python
# A. Crear un diccionario vacío
traductor = {}

# B. Añadir datos (Sintaxis: diccionario[llave] = valor)
traductor["apple"] = "manzana"
traductor["blue"] = "azul"

# C. También puedes crearlo con datos iniciales
precios = {"pan": 30, "leche": 120}

# D. Añadir un nuevo dato a 'precios'
precios["huevo"] = 15

print("Traductor:", traductor)
print("Precios actualizados:", precios)
```

```python
## OPERACIONES
# A. Crear y añadir/actualizar
auto = {"marca": "Toyota", "año": 2020}
auto["color"] = "Rojo"       # Añadir nueva llave
auto["año"] = 2024           # Actualizar valor de llave existente

# B. Acceso seguro
modelo = auto.get("modelo", "No especificado") # Evita error si la llave no existe

# C. Extraer llaves y valores
todas_las_llaves = list(auto.keys())
todos_los_valores = list(auto.values())
pares_completos = list(auto.items())  # Devuelve lista de tuplas (llave, valor)

# D. Eliminar
del auto["marca"]            # Elimina la llave y su valor
auto.clear()                 # Vacía todo el diccionario

print("Llaves del auto:", todas_las_llaves)
```

### Detalles de como funiona a nivel físico


Para entender la diapositiva del profesor Takai (Página 31) a un nivel avanzado, debemos ver la relación entre el **Software** (el código que escribes) y el **Hardware** (los transistores dentro del chip).

Lo que ves en pantalla es el ciclo de ejecución de una **CPU Síncrona**. Aquí está el detalle técnico de lo que ocurre en cada parte del gráfico:

---

##### 1. El Concepto de Direccionamiento (Memory Address)

El texto japonés arriba del recuadro azul dice: "El programa se ejecuta en orden de direcciones de memoria pequeñas a grandes".

* **La Memoria RAM** es como un hotel con miles de habitaciones numeradas.
* Tu código (`a = 10`, `print(a)`) se traduce a código binario (ceros y unos) y se guarda en esas habitaciones: la dirección `0000`, luego la `0001`, etc.
* La CPU siempre sabe dónde está parada gracias a un registro llamado **Program Counter (PC)**.

---

##### 2. El Reloj de la CPU (La onda cuadrada)

Ese dibujo con picos altos y bajos es la señal de reloj.

* **Reloj (Clock):** Es un cristal que vibra a una velocidad constante (en el ejemplo, 2.8GHz, lo que significa 2,800 millones de veces por segundo).
* **Flanco de Subida (Rising Edge):** Las flechas hacia arriba indican que la CPU solo realiza acciones cuando el pulso eléctrico sube. Es el "latido del corazón" que sincroniza todo.

---

##### 3. El Ciclo de Instrucción (Fetch - Execute)

El gráfico divide la operación en dos fases que se repiten infinitamente:

###### **Fase A: Fetch (フェッチ - Buscar)**

1. La CPU envía la dirección actual del **Program Counter** (ej. `0000`) a la memoria.
2. La memoria le devuelve el "código de operación" (la instrucción binaria) que está ahí guardada.
3. Este código llega a la CPU y se guarda en un registro de instrucción.

###### **Fase B: Execute (実行 - Ejecutar)**

1. La CPU "entiende" qué significa ese código (ej. "sumar" o "asignar valor").
2. Si es un cálculo (`b = a * 1.1`), usa la **ALU (Unidad Aritmético Lógica)**.
3. **Crucial:** Mientras ejecuta, el **Program Counter** suma `+1`. Esto garantiza que, en el próximo latido del reloj, la CPU busque en la dirección `0001`.

---

##### 4. Datos Técnicos de la Diapositiva

* **Intel Xeon 2.8GHz:** Es un procesador de servidor.
* **360ps (Picosegundos):** Es el tiempo que dura un solo ciclo de reloj. Como referencia, la luz recorre solo unos 10 centímetros en 360 picosegundos. ¡Esa es la ventana de tiempo que tiene el procesador para mover electricidad de un punto a otro!
* **Quad-Core:** El profesor aclara que, aunque el ejemplo es simple, en la realidad hay 4 CPUs (núcleos) trabajando en paralelo, lo que hace que la coordinación sea más compleja.

---

##### Resumen:

> "La CPU ejecuta las instrucciones de forma sincronizada por un **reloj**. En cada pulso, realiza un **Fetch** (traer código de la memoria) y un **Execute** (realizar la acción), avanzando automáticamente a la siguiente dirección de memoria mediante el **Program Counter**."

##### Problemas
1. El Problema del "Timing" (Carrera de Datos)
La electricidad no viaja de forma instantánea. Tarda unos picosegundos en llegar desde la RAM hasta los pines de la CPU.

El reloj está diseñado para esperar el tiempo justo a que los voltajes se estabilicen.

Si la CPU intenta hacer el Execute antes de que el Fetch haya terminado de "llenar la tubería" de electricidad, la CPU leerá basura (un estado indeterminado entre 0 y 1).

Resultado: El programa ejecuta una instrucción que ni siquiera existe.

2. El Caos del Program Counter (PC)
Como viste en la diapositiva, el Program Counter suma +1 en cada ciclo.

Si hay un "glitch" o desincronización, el PC podría saltar de la dirección 0001 a la 0005 por error, o quedarse atrapado en la 0001 para siempre.

Es como si estuvieras leyendo un libro y, de repente, el viento pasara cinco páginas a la vez. Perderías el hilo de la historia (el hilo de ejecución).

3. "Setup and Hold" (Estabilidad de la señal)
A nivel de transistores, para que un dato se quede grabado en un registro de la CPU, la señal debe estar estable un tiempo antes y un tiempo después del pulso de reloj (esto se llama Setup time y Hold time).

Si el reloj llega tarde o temprano, el transistor no llega a cerrarse del todo.

El bit se queda en un estado llamado Metastabilidad (no es ni 0 ni 1).

Resultado: La CPU se "congela" (el famoso freeze) porque los circuitos lógicos entran en un estado de contradicción.

En resumen:
La CPU es como un director de orquesta que marca el ritmo con su batuta (el reloj).

Los músicos (ALU, Registros, Memoria) son extremadamente rápidos pero no saben qué nota sigue por sí mismos.

Si el director pierde el ritmo, el violinista empieza el solo antes de que el pianista termine la introducción.

En un ordenador, ese "ruido musical" se traduce en un Crash, una Blue Screen o un Kernel Panic.

## **Programación de Bifurcación (Branching / 分岐)**
**Lógica:** "Si la condición es verdadera, haz A; si no, haz B".
**Palabras clave:** `if`, `elif`, `else`.

*Nota: ojo con la indentación*

### **`if`, `elif`, `else` basics**

`if` solo: Es una decisión simple (Vas por el camino A o sigues derecho).

`if-else`: Es un camino que se divide en dos (Vas por A o vas por B, pero no por ambos).

`if-elif-else`: Es un camino que se divide en muchas opciones, pero al final solo puedes elegir una sola.

Con el `elif`, en cuanto Python encuentra una respuesta verdadera, deja de preguntar por las demás, ahorrando energía a la CPU.

```python
a = 10

if a == 10:
    print("aaa")  # Se ejecuta porque la condición es cierta

print("bbb")      # Se ejecuta siempre (está fuera del if)

a = 100

if a == 1:
    print("aaa")  # Python IGNORA esta línea porque 10 no es 1

print("bbb")      # Se ejecuta siempre (es el flujo secuencial principal)
```

### **Truthy & Falsy**
"Expresiones condicionales un poco extrañas" (ちょっと変わった条件式)



**Falsy** (Actúan como False): El número `0`, un texto vacío `""`, una lista vacía `[]`, o `None`.

**Truthy** (Actúan como True): Cualquier número que no sea 0 (como `1, -5, 0.1)` y cualquier texto que tenga al menos un carácter (como `" ", "a", "0"`).

```python
# [2]
a = True
print(a)
if a:
    print("aaa")

# [3]
a = False
print(a)
if a:
    print("aaa")

# [4]
a = 0
print(a)
if a:
    print("aaa")

# [5]
a = 1
print(a)
if a:
    print("aaa")

# [6]
a = ""
print(a)
if a:
    print("aaa")

# Ejemplo final
a = "he"
print(a)
if a:
    print("aaa")
```

```python
letrasdeminombre = ["s","o","f","i","a"]
vocales = ["a","e","i","o","u"]

for i in letrasdeminombre:
    if i == vocales:
        print(i)
    elif i in vocales:
        print(i)

print("*"*20)

nombre = "sofia"
vocales = "aeiou"

for letra in nombre:
    if letra in vocales:
        print(letra)

print("*"*20)

set(letrasdeminombre) #esto por si solo NO GUARDA la variable como set, solo crea una nueva pero no la referencia de nuevo
set(vocales) #por eso se debe hacer esto:

set_letrasdeminombre = set(letrasdeminombre)
set_vocales = set(vocales)

comunes = set_letrasdeminombre & set_vocales
print(comunes)

print("*"*20)

comunes = set(letrasdeminombre) & set(vocales)
print(comunes)
```

#### **`range(start, stop, step)`**
`start (Inicio):` El primer número que aparece. Si no pones nada, Python asume que es 0.

`stop (Límite):` El bucle se detiene justo antes de llegar a este número. Nunca lo toca.

`step (Paso):` Es el salto entre números, determina la "velocidad" o el intervalo del conteo.. Puede ser hacia adelante (positivo) o hacia atrás (negativo).

Si el step es 2, cuentas de dos en dos (2, 4, 6...).

Si el step es 10, saltas secciones grandes (10, 20, 30...).

Si el step es -1, ¡puedes contar hacia atrás! (10, 9, 8...).

```python
# ---
print("Ejemplo 1 (solo stop):")
for i in range(5):
    print(i, end=" ") # El end=" " es para que salgan en la misma línea

# ---
print("\n\nEjemplo 2 (start y stop):")
for i in range(5, 12):
    print(i, end=" ")

# ---
print("\n\nEjemplo 3 (start, stop y step):")
for i in range(2, 21, 2):
    print(i, end=" ")

# ---
print("\n\nEjemplo 4 (start, stop y step negativo):")
for i in range(10, 0, -1):
    print(i, end=" ")
```

```python
letras = ["A", "B", "C"]

for letra in letras:              # Bucle Externo
    for numero in range(1, 4):    # Bucle Interno
        print(f"Código generado: {letra}{numero}")
```

```python
from ColabTurtlePlus.Turtle import *
import random

clearscreen()
setup(400, 400)
t = Turtle()
t.speed(10)

tamaño = 10 # Empezamos con un radio pequeño

for angulo in range(0, 360, 30): # Bucle "Padre" (Posiciones)
    t.penup()
    t.forward(80)
    t.pendown()
    color_interior = t.pencolor((random.randint(0,255), random.randint(0,255), random.randint(0,255)))

    # Bucle "Hijo" (Dibuja 3 círculos uno dentro de otro en esa posición)
    for radio_pequeño in range(10, 40, 10):
        t.circle(radio_pequeño)
        t.pencolor(color_interior)

    t.penup()
    t.goto(0,0)
    t.left(30)

    t.penup()
    t.goto(0,0)
    t.left(30)

saveSVG()
```

```python
from ColabTurtlePlus.Turtle import *
import random

# 1. Configuración del lienzo
clearscreen()
setup(500, 500)
t = Turtle()
t.speed(13) # Velocidad máxima

# Tamaño de cada celda de la cuadrícula
tamaño_celda = 40

# --- ESTRUCTURA DE BUCELES ANIDADOS (NESTING) ---

# BUCLE EXTERNO: Controla las 5 filas (eje Y)
for fila in range(5):
    # BUCLE MEDIO: Controla las 5 columnas por fila (eje X)
    for columna in range(5):

        t.pencolor((random.randint(0,255), random.randint(0,255), random.randint(0,255)))

        # 1. Calcular la posición exacta de esta celda
        # Empezamos en (-100, -100) para centrar la cuadrícula un poco
        x = -100 + (columna * tamaño_celda)
        y = -100 + (fila * tamaño_celda)

        # 2. Mover la tortuga sin dibujar
        t.penup()
        t.goto(x, y)
        t.pendown()

        # 3. Elegir color aleatorio para el relleno de esta celda
        t.fillcolor((random.randint(0,255), random.randint(0,255), random.randint(0,255)))

        # 4. BUCLE INTERNO: Dibujar y rellenar UN cuadrado (celda)
        t.begin_fill()
        for i in range(4): # 4 lados
            t.forward(tamaño_celda)
            t.right(90) # Girar 90 grados
        t.end_fill()

# Guardar el dibujo final
saveSVG()
```
